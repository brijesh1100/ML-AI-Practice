```python
import os, json
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
```

```python
df = pd.read_csv("./Inferential_Statistics_Student_Loan.csv")
df.head()
```

```python
result_df = df.drop(['Unnamed: 4','Unnamed: 5'], axis=1)
```

```python
round(100*(result_df.isnull().sum()/len(result_df.index)), 2)
```

```python
result_df.head()
result_df.info()
```

```python
result_df=result_df.rename(index=str, columns={"Exposure at Default (in lakh Rs.)" : "EAD", "Recovery (%)":"Recovery_in_percentage", "Probability of Default":"Probability_of_Default"} )
result_df.head()
```

```python
result_df['Recovery_in_percentage'] = result_df['Recovery_in_percentage'].map(lambda x : x.rstrip('%')).astype(float)
result_df.head()
```

```python
result_df['X_Money_Lost_by_Bank_In_Lakh']=(result_df.EAD * result_df.Recovery_in_percentage)/100
result_df.head()
```

```python
result_df['Expected_loss_in_Lakh']= result_df.Probability_of_Default*result_df.EAD*(1-(result_df.Recovery_in_percentage/100))
```

```python
result_df.head()
```

```python
result_df['expected_loss_in_RS'] = result_df.Expected_loss_in_Lakh * 100000
result_df.head()
```

```python
Total_loss = result_df['expected_loss_in_RS'].sum()
print('Total Loss in cr :', Total_loss/10000000)
```
